## Importing Files

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.model_selection import GridSearchCV
```

## 1. Problem Statement

The objective is to find out some relation between features of a mobile phone(eg:- RAM,
Internal Memory, etc) and its selling price. In this problem, we do not have to predict the
actual price but a price range indicating how high the price is.

### 4. Feature Engineering

```python
x = df_mobile.drop('price_range',axis=1)
y = df_mobile['price_range']
```

### Scaling - Normalization

```python
normalization = MinMaxScaler()
array = normalization.fit_transform(x)
norm_scaling = pd.DataFrame(array,columns=x.columns)
norm_scaling
```

### 5. Feature Selection

### 6. Model Training

```python
x_train,x_test,y_train,y_test = train_test_split(norm_scaling,y,train_size=0.8,random_state=25,stratify=y)
```

### 7. Model Selection

```python
knnclass = KNeighborsClassifier()
knnclass.fit(x_train,y_train)
```

### Evaluation - Training Dataset

```python
y_pred_train = knnclass.predict(x_train)

cnf1        = confusion_matrix(y_train,y_pred_train)
accuracy1   = accuracy_score(y_train,y_pred_train)
clf_report1 = classification_report(y_train,y_pred_train)

print('Confusion Matrix :\n',cnf1,'\n')
print('Accuracy :',accuracy1,'\n')
print('Classification Report :\n',clf_report1)
```

### Evaluation - Testing Dataset

```python
y_pred_test = knnclass.predict(x_test)

cnf        = confusion_matrix(y_test,y_pred_test)
accuracy   = accuracy_score(y_test,y_pred_test)
clf_report = classification_report(y_test,y_pred_test)

print('Confusion Matrix :\n',cnf,'\n')
print('Accuracy :',accuracy,'\n')
print('Classification Report :\n',clf_report)
```

### Hyper Parameter Tunning - Grid SearchCV

```python
hyperparameter = {'n_neighbors':np.arange(3,40),
                 'p':[1,2]}
gs = GridSearchCV(knnclass,hyperparameter,cv=5)
gs.fit(x_train,y_train)
gs.best_estimator_
```

```python
knnclf = KNeighborsClassifier(n_neighbors=37, p=1)
knnclf.fit(x_train,y_train)
```

```python
# Evaluation - Training

y_pred_train = knnclf.predict(x_train)

cnf1        = confusion_matrix(y_train,y_pred_train)
accuracy1   = accuracy_score(y_train,y_pred_train)
clf_report1 = classification_report(y_train,y_pred_train)

print('Confusion Matrix :\n',cnf1,'\n')
print('Accuracy :',accuracy1,'\n')
print('Classification Report :\n',clf_report1)
```

```python
# Evaluation - Testing

y_pred_test = knnclf.predict(x_test)

cnf        = confusion_matrix(y_test,y_pred_test)
accuracy   = accuracy_score(y_test,y_pred_test)
clf_report = classification_report(y_test,y_pred_test)

print('Confusion Matrix :\n',cnf,'\n')
print('Accuracy :',accuracy,'\n')
print('Classification Report :\n',clf_report)
```
